```python
#All import statements can go up here
from joblib import Memory

from astroquery.gaia import Gaia

import numpy as np
from numpy.lib.recfunctions import structured_to_unstructured

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.colors as clr

import scipy as sci

import astropy as ast
from astropy.table import Table
import astropy.units as u

import pandas as pd

import time
```

```python
#Copy slightly better caching function
cachedir = './joblib_cache'
memory = Memory(cachedir, verbose=0)

@memory.cache
def get_gaia_query(q, unlimited=True):
    start = time.time()
    if unlimited:
        Gaia.ROW_LIMIT = -1
    job = Gaia.launch_job_async(q)
    print(f"Total time: {time.time()-start:0.2f} sec")
    return job.get_results()
```

```python
#Get results for query developed on gaia database
query = """SELECT TOP 100 * 
FROM gaiadr3.vari_rrlyrae
WHERE pf IS NOT NULL
AND num_clean_epochs_g > 40
ORDER BY num_clean_epochs_g DESC"""
results = get_gaia_query(query, unlimited=False)
```
