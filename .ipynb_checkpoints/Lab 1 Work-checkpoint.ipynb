## Query Caching System and Query 1 (step 1)

```python
#Copy slightly better caching function
cachedir = './joblib_cache'
memory = Memory(cachedir, verbose=0)

@memory.cache
def get_gaia_query(q, unlimited=True):
    start = time.time()
    if unlimited:
        Gaia.ROW_LIMIT = -1
    job = Gaia.launch_job_async(q)
    print(f"Total time: {time.time()-start:0.2f} sec")
    return job.get_results()
```

```python
#Get results for query developed on gaia database
query = """SELECT TOP 100 * 
FROM gaiadr3.vari_rrlyrae
WHERE pf IS NOT NULL
AND num_clean_epochs_g > 40
ORDER BY num_clean_epochs_g DESC"""
results = get_gaia_query(query, unlimited=False)
```

```python
# 1. Define the source ID (Example chosen is id of the star in results with the most clean epochs observed)
source_id = 5284173207165634048

# 2. Query DataLink for the file URLs
data_release='Gaia DR3'
retrieval_type='EPOCH_PHOTOMETRY' # We ask for EPOCH_PHOTOMETRY to get the time series data
data_structure='INDIVIDUAL'
linking_parameter='SOURCE_ID'
file_format='xml' # Defaulting to the 'votable' format in astroquery
datalink = Gaia.load_data(ids=[source_id],
                          data_release=data_release, 
                          retrieval_type=retrieval_type, 
                          data_structure=data_structure,
                          linking_parameter=linking_parameter)

# Here we check the names of the datalink files we have accessed.
print(f'The following Datalink products have been downloaded:')
for dl_key in datalink.keys():
    print(f' * {dl_key}')

# 3. Extract and combine the timeseries data for the desired source.
source_filename = f'{retrieval_type}-{data_release} {source_id}.{file_format}'
data = table.vstack([dl.to_table() for dl in datalink[source_filename]])
# For other file_format parameters (e.g. 'csv', 'fits'), the above line can be simplified to 'data = table.vstack(datalink[source_filename])'

# 4. Plotting the raw time series
plt.figure(figsize=(10,4))

time = data['g_transit_time']
g_band_mag = data['g_transit_mag']
plt.scatter(time, g_band_mag, s=10, color='green')
plt.gca().invert_yaxis() # Magnitudes are inverted!
plt.xlabel(f'Time [{time.unit}]')
plt.ylabel(f'G [{g_band_mag.unit}]')
plt.title(f"RR Lyrae Light Curve (Source {source_id})")
plt.show()
```

## My Epoch Photometry (effectively repeat querying above for all 100 stars (and cache results?)) (Step 2)

```python
#clear cache (when needed)
def clear_cache():
    memory.clear(warn=False)
#clear_cache()
```

```python
#Function decorator will recache results when called with the same arguments, SO:
#Caching one star at a time instead of all of them at once protects against recaching all of the stars (and wasting data)
#If only one of them is actually different from before.
@memory.cache
def cache_a_star(source_id=0, data_release="Gaia DR3", retrieval_type='EPOCH_PHOTOMETRY', data_structure='INDIVIDUAL', linking_parameter='SOURCE_ID'):
    data = Gaia.load_data(ids=[source_id],
                          data_release=data_release, 
                          retrieval_type=retrieval_type, 
                          data_structure=data_structure,
                          linking_parameter=linking_parameter)
    return {
        key: table.vstack([d.to_table() for d in value])
        for key, value in data.items()
    }

total = 0
datalinks = []
for star_id in results['source_id']:
    datalinks.append(cache_a_star(star_id))
    total += 1
    if total % 10 == 0:
        print(total, "Stars' Lightcurves Downloaded!")

print(f'The following Datalink products have been downloaded:')
for datalink in datalinks:
    for dl_key in datalink.keys():
        print(f' * {dl_key}')
```

## Lomb-Scargle for the 2 Stars Above Without Including Error Quantities (Step 3)

```python
def plot_lomb_scargle(source_id=5284173207165634048, w=10, h=5, c="red"):
    datalink = cache_a_star(source_id) #Call this to get back our previously cached file
    source_filename = f'{retrieval_type}-{data_release} {source_id}.{file_format}'
    data = datalink[source_filename]

    t = data['g_transit_time'].value #remove units to dodge errors
    y = data['g_transit_mag']

    #remove data missing values so we don't get NaN errors
    mask = np.isfinite(t) & np.isfinite(y) 
    t = t[mask]
    y = y[mask]

    flux = 10**(-0.4 * y) #Use flux, not just magnitudes for better result -> magnitudes are logarithmic, flux is more objective (although it also makes outliers stand out a lot more)

    freq, p_ls = LombScargle(
        t, flux,
        fit_mean=True,
        center_data=True
    ).autopower(
        minimum_frequency=1.0, #RR lyrae tends to be specifically in this range (1-5 brightness cycles per day)
        maximum_frequency=5.0,
        samples_per_peak=10
    )

    fig, ax = plt.subplots(1, figsize=(w, h))
    ax.plot(freq, p_ls, label='Lomb-Scargle Periodogram', color=c)
    ax.legend();
    ax.set(title=f"Lomb-Scargle Periodogram of Source ID {source_id} (Excluding Errorbars)")
    ax.set(ylabel='Spectral Power')
    ax.set(xlabel='frequency')

plot_lomb_scargle(source_id=5284173207165634048, w=12, h=6)
plot_lomb_scargle(source_id=5284160459706483200, w=12, h=6, c="green")
```

## Lomb-Scargle Including Error Bars (Step 3)

```python
def mag_error(flux, flux_error, zp_error=0.003):
    m_flux = 1.0857 * (flux_error / flux)
    total_error = np.sqrt(m_flux ** 2 + zp_error ** 2)
    return total_error
```

```python
def plot_lomb_scargle_with_errors(source_id=5284173207165634048, w=10, h=5, c="red", markerline=-1, mc="white"):
    datalink = cache_a_star(source_id) #Call this to get back our previously cached file
    source_filename = f'{retrieval_type}-{data_release} {source_id}.{file_format}'
    data = datalink[source_filename]

    t = data['g_transit_time'].value #remove units to dodge errors
    y = data['g_transit_mag']

    #remove data missing values so we don't get NaN errors
    mask = np.isfinite(t) & np.isfinite(y) 
    t = t[mask]
    y = y[mask]
    data = data[mask] #So I can assign the right errors to each term with mag_error

    flux = 10**(-0.4 * y) #Use flux, not just magnitudes for better result -> magnitudes are logarithmic, flux is more objective (although it also makes outliers stand out a lot more)
    absMag = -2.5 * np.log10(flux) #Flux-weighted g-mag

    magErrors = mag_error(flux, data["g_transit_flux_error"].value)

    freq, p_ls = LombScargle(
        t, absMag, magErrors,
        fit_mean=True,
        center_data=True
    ).autopower(
        minimum_frequency=1.0, #RR lyrae tends to be specifically in this range (1-5 brightness cycles per day)
        maximum_frequency=5.0,
        samples_per_peak=10
    )

    fig, ax = plt.subplots(1, figsize=(w, h))
    ax.plot(freq, p_ls, label='Lomb-Scargle Periodogram', color=c)

    if markerline != -1: #make using marker optional
        plt.axvline(x=markerline, color=mc, linestyle='--', label=f'Manually Estimated Period: {markerline:.4f} days')
        meanPer = freq[np.argmax(p_ls)]
        plt.axvline(x=meanPer, color='blue', linestyle=':', label=f'Program-Determined Period: {meanPer:.4f} days')
    
    ax.legend();
    ax.set(title=f"Lomb-Scargle Periodogram of Source ID {source_id} (Including Errorbars)")
    ax.set(ylabel='Spectral Power')
    ax.set(xlabel='frequency')

plot_lomb_scargle_with_errors(source_id=5284173207165634048, w=12, h=6, markerline=1.48)
plot_lomb_scargle_with_errors(source_id=5284160459706483200, w=12, h=6, c="green", markerline=1.685)
```

The frequency of the bottom is larger, closer to ~1.685, so period must be 1/1.685 or ~0.5935 days

## Calculating the Other 98 (Step 3 Again)

```python
def do_ls_with_error(source_id=5284173207165634048):
    datalink = cache_a_star(source_id) #Call this to get back our previously cached file
    source_filename = f'{retrieval_type}-{data_release} {source_id}.{file_format}'
    data = datalink[source_filename]

    t = data['g_transit_time'].value #remove units to dodge errors
    y = data['g_transit_mag']

    #remove data missing values so we don't get NaN errors
    mask = np.isfinite(t) & np.isfinite(y) 
    t = t[mask]
    y = y[mask]
    data = data[mask] #So I can assign the right errors to each term with mag_error

    flux = 10**(-0.4 * y) #Use flux, not just magnitudes for better result -> magnitudes are logarithmic, flux is more objective (although it also makes outliers stand out a lot more)
    absMag = -2.5 * np.log10(flux) #Flux-weighted g-mag

    magErrors = mag_error(flux, data["g_transit_flux_error"].value)

    freq, p_ls = LombScargle(
        t, absMag, magErrors,
        fit_mean=True,
        center_data=True
    ).autopower(
        minimum_frequency=1.0, #RR lyrae tends to be specifically in this range (1-5 brightness cycles per day)
        maximum_frequency=5.0,
        samples_per_peak=10
    )
    return 1/(freq[np.argmax(p_ls)]) #return frequency where spectral power is largest, by taking the inverse of the most likely period

print(do_ls_with_error()) #For top star above
print(do_ls_with_error(source_id=5284160459706483200)) #For bottom star above
```

```python
#Now, just call that function on all stars for a list of all of their periods
all_ests = []
for star in results:
    star_id = star['source_id']
    all_ests.append(do_ls_with_error(star_id))

print(len(all_ests)) #confirm all 100 get values
print(all_ests)
```

## Step 5 (Just A Bunch of LaTeX Markdown Math)

If the angular frequency $\omega = 2\pi/P$ is known, then the Fourier model

$$
f(t) = A_0 + \sum_{k=1}^{K} \left[ a_k \sin(k\omega t) + b_k \cos(k\omega t) \right]
$$

is linear in the unknown coefficients $A_0, a_k, b_k$.

Once $\omega$ is fixed, all sine and cosine terms are just known numbers at each time $t$. 
The only unknowns are the coefficients in front of them.

### Write the model at measured times

Suppose we measure the flux at times $t_1, t_2, \dots, t_N$.
Let

$$
y_i = f(t_i)
$$

Then

$$
y_i = A_0 + \sum_{k=1}^{K} \left[ a_k \sin(k\omega t_i) + b_k \cos(k\omega t_i) \right]
$$

### Write one observation as a dot product

Define the vector of known quantities:

$$
X_i =
\begin{bmatrix}
1 \\
\sin(\omega t_i) \\
\cos(\omega t_i) \\
\sin(2\omega t_i) \\
\cos(2\omega t_i) \\
\vdots \\
\sin(K\omega t_i) \\
\cos(K\omega t_i)
\end{bmatrix}
$$

Define the parameter vector:

$$
\beta =
\begin{bmatrix}
A_0 \\
a_1 \\
b_1 \\
a_2 \\
b_2 \\
\vdots \\
a_K \\
b_K
\end{bmatrix}
$$

Then

$$
y_i = X_i^T \beta
$$

### Stack all observations

Stack all observations into

$$
y =
\begin{bmatrix}
y_1 \\
y_2 \\
\vdots \\
y_N
\end{bmatrix}
$$

Construct the design matrix

$$
X =
\begin{bmatrix}
1 & \sin(\omega t_1) & \cos(\omega t_1) & \cdots & \sin(K\omega t_1) & \cos(K\omega t_1) \\
1 & \sin(\omega t_2) & \cos(\omega t_2) & \cdots & \sin(K\omega t_2) & \cos(K\omega t_2) \\
\vdots & \vdots & \vdots & & \vdots & \vdots \\
1 & \sin(\omega t_N) & \cos(\omega t_N) & \cdots & \sin(K\omega t_N) & \cos(K\omega t_N)
\end{bmatrix}
$$

### Linear Form

The full problem becomes

$$
y = X \beta
$$

Here:

- $y$ is $N \times 1$  
- $X$ is $N \times (2K+1)$  
- $\beta$ is $(2K+1) \times 1$  

Since $\omega$ is known, every entry of $X$ is known. 
The only unknowns are the coefficients in $\beta$.
Therefore this is a standard linear least-squares problem:

$$
\hat{\beta} = (X^T X)^{-1} X^T y
$$


## Step 6: Plot A Phase-Folded Light Curve and Lines of Best Fit
K-Values that matter: 1, 3, 5, 7, & 9

```python
#Plot the Phase-Folded Lightcurve
plot_phase_folded_lightcurve(source_id=4659759557323962752, cscheme="Reds_r")
```

```python
#Plot everything together function
def plot_curve_and_lines(source_id=4659759557323962752, k_range=[1,3,5,7,9], w=13, h=6, colors=['blue', 'blue', 'blue', 'blue', 'blue'], lws=[3,3,3,3,3], linestyles=['-', '-', '-', '-', '-']):
    plot_phase_folded_lightcurve(source_id=source_id, cscheme="Reds_r", autoShow=False, w=w, h=h)

    #Call this to get back our previously cached file
    datalink = cache_a_star(source_id)
    source_filename = f'{retrieval_type}-{data_release} {source_id}.{file_format}'
    data = datalink[source_filename]

    starRow = results[results['source_id'] == source_id]
    
    times = data['g_transit_time'].value
    mags = data['g_transit_mag'].value
    period = float(starRow["pf"].value)

    phases = (times % period) / period
    thetas = 2 * np.pi * phases

    j = 0
    for k in k_range:

        #Get matrix values
        
        kMatrix = np.ones((len(times), 2*k+1))
        for i in range(1, k+1):
            kMatrix[:, 2*i-1] = np.sin(i * thetas)
            kMatrix[:, 2*i] = np.cos(i * thetas)

        coefficients, residuals, rank, s = np.linalg.lstsq(kMatrix, mags, rcond=None)

        #Overplot line
        phaseGrid = np.linspace(0,1,5000) #ultra-fine
        thetaGrid = 2 * np.pi * phaseGrid

        kGrid = np.ones((len(phaseGrid), 2*k + 1))
        for i in range(1, k+1):
            kGrid[:, 2*i-1] = np.sin(i * thetaGrid)
            kGrid[:, 2*i] = np.cos(i * thetaGrid)
        
        fit = kGrid @ coefficients
        
        plt.plot(phaseGrid, fit, label=f'Fitted Series At K={k}', color=colors[j], linewidth=lws[j], linestyle=linestyles[j])
        j += 1

    plt.legend()
    plt.title(f"Phase-Folded Lightcurve of Source ID {source_id} With Overplotted Range of Fourier Series")
    plt.show()

plot_curve_and_lines(w=19.5, h=9, colors=['darkblue', 'blue', 'dodgerblue', 'deepskyblue', 'cyan'], lws=[3,3,3,3,4.5], linestyles=[':', ':', ":", ":", "-"]) #lighter colors representing higher levels, thicker solid line + lightest hue for best estimating K.
```
